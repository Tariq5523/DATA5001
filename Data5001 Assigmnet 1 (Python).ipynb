```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
import seaborn.objects as so
```

```python
# For generating random data
rng = np.random.default_rng(3)
```

```python
# Generating data and adding them to a dataframe
x = rng.standard_normal(100)
y = rng.standard_normal(100)
df1 = pd.DataFrame({'x': x, 'y': y})
df1.head()
```

```python
# Change the these of all seaborn plots
sns.set_theme()
```

```python
# Creating a lineplot using seaborn
sns.lineplot(data=df1, x="x", y="y")
```

```python
# Creating a scatter plot using seaborn
sns.scatterplot(data=df1, x='x', y='y').set_title("Plot of X VS Y")
```

```python
# Creating more than one plot using seaborn
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
sns.scatterplot(data=df1, x="x", y="y",color='coral', ax=axes[0,1])
sns.scatterplot(data=df1, x="x", y="y",marker="+",color='black', ax=axes[1,2])
```

# Could not contour plots using seaboen

```python
# contour plots
fig, ax = plt.subplots(figsize=(8, 8))
x = np.linspace(-np.pi, np.pi, 50)
y = x
f = np.multiply.outer(np.cos(y), 1 / (1 + x**2))
ax.contour(x, y, f, levels=45);
```

```python
# contour plots using imshow()
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(f);
```

## Graphs from the "Additional Graphical and Numerical Summaries" section in the lab

```python
# Reading data in a pandas dataset
Auto = pd.read_csv('Auto.csv', na_values=['?'])
```

```python
# Converting cylinders colmn to a categorical type for graphing later
Auto.cylinders = pd.Series(Auto.cylinders, dtype='category')
Auto.cylinders.dtype
```

```python
# Creating a scatter plot using seaborn and Auto data (horsepower vs mpg)
sns.scatterplot(data=Auto, x='horsepower', y='mpg',color='coral')
plt.title("Horsepower VS MPG")
```

```python
# Creating a grid plot using seabon and Auto data (horsepower vs mpg)
fig, axes = plt.subplots(ncols=3, figsize=(15, 5))
sns.scatterplot(data=Auto, x='horsepower', y='mpg',color='coral',ax=axes[1]);
```

```python
# Creating a boxplot using seaborn and Auto data
sns.boxplot(data=Auto, x='cylinders', y='mpg')
plt.title('MPG Grouped by Cylinders')
```

```python
# Creating a histogram using seaborn and Auto data
sns.histplot(data=Auto, x='mpg')
plt.title("MPG Histogram")
```

```python
# Change theme of graph to show ticks on x and y axes
sns.set_theme(style="ticks")
# Creating a scatter matrix using seaborn and Auto data
sns.pairplot(Auto)
```

```python
# Creating a scatter matrix using sea born for a subset of the Auto data
sns.pairplot(Auto[['mpg','displacement','weight']])
```
